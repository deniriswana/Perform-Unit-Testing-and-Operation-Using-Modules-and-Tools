### datetime
- A date in Python is not a data type of its own, but we can import a module named datetime to work with dates as date objects.

```python
import datetime
```

```python
todayWithTime = datetime.datetime.today()
```

```python
todayWithTime
```

```python
todayWithoutTime = datetime.date.today()
```

```python
todayWithoutTime
```

```python
formatting_time = datetime.datetime.strftime(todayWithTime, "%H: %M: %S")
```

```python
formatting_time
```

```python
formatting_date = datetime.datetime.strftime(todayWithoutTime, "%m/%d/%y")
```

```python
formatting_date
```

### The strftime() method
- The datetime object has a method for formatting date objects into readable strings.
-The method is called strftime(), and takes one parameter, format, to specify the format of the returned string:

```python
import datetime as date
 
birthday = datetime.date(2000, 2, 21)
 
print(birthday.strftime("%A, %d %B %Y"))
```
