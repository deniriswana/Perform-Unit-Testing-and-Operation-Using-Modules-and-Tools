- formula for the circumference of a circle
- formula for the area of a circle

```python
pi = math.pi
rad = int(input("Enter the radius of the circle: "))

luasL = math.pi * rad * rad
kelL = math.pi * 2 * rad

print("The area of circle is", luasL)
print("The circumference of circle is", kelL)
```
